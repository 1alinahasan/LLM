```python
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer

# Load the GPT-J model and tokenizer
gpt_j_model = AutoModelForCausalLM.from_pretrained("EleutherAI/gpt-j-6B")
gpt_j_tokenizer = AutoTokenizer.from_pretrained("EleutherAI/gpt-j-6B")

# Initialize the text generation pipeline with GPT-J
text_generation_pipeline = pipeline(
    "text-generation",
    model=gpt_j_model,
    tokenizer=gpt_j_tokenizer
)

def generate_response(question):
    # Generate a response using the GPT-J pipeline
    conversation_history = f"Human: {question}\nAI:"
    response = text_generation_pipeline(
        conversation_history,
        max_length=100,
        truncation=True,
        pad_token_id=text_generation_pipeline.tokenizer.eos_token_id
    )

    # Extract and return the generated response
    generated_text = response[0]['generated_text'].strip()
    return generated_text

# Ask GPT-J about the capital of France
question = "What is the capital of France?"
response = generate_response(question)

# Print the response
print("GPT-J Response:", response)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Initialize GPT-J model and tokenizer
model_name = "EleutherAI/gpt-j-6B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Initialize the text generation pipeline with GPT-J
text_generation_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer)

def generate_response(question):
    # Generate a response using GPT-J pipeline
    conversation_history = f"Human: {question}\nAI:"
    response = text_generation_pipeline(
        conversation_history,
        max_length=100,
        truncation=True,
        pad_token_id=tokenizer.eos_token_id
    )

    # Extract and print the generated response
    generated_text = response[0]['generated_text'].strip()
    print(f"Q: {question}")
    print(f"A: {generated_text}\n")

# List of common QK questions
questions = [
    "Who was the first person to walk on the moon?",
    "What is the capital of France?",
    "How do plants perform photosynthesis?",
    "What are the three laws of motion?",
    "Can you explain the theory of relativity in simple terms?",
]

# Loop through the questions and generate responses
for question in questions:
    generate_response(question)
```
